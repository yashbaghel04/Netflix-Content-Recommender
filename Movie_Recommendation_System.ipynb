Removing the space between the names of one person

```python
def no_space(text):
  if isinstance(text,str):
    no_space =text.split(" ")
    return "".join(no_space)
    return ''
df['cast']= df['cast'].apply(no_space)
```

Reducing the cast size to only 3 person

```python
def get_three(text):
  if isinstance(text, str):


    cast_3 = [actor.strip() for actor in text.split(',') if actor.strip()]
    return ", ".join(cast_3[:3])
  return ''

df["cast"] = df["cast"].apply(get_three)
```

```python
import nltk
from  nltk.stem.porter import PorterStemmer
ps =PorterStemmer()
```

```python
def stem(text):
  l = []
  for i in text.split(" "):
    l.append(ps.stem(i))
  return " ".join(l)

movies.loc[:, 'tags'] = movies['tags'].apply(stem)
```

```python
movies['tags'][1]
```

```python
v =CountVectorizer(max_features=8807,stop_words='english')
vector=v.fit_transform(movies['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar =cosine_similarity(vector)
```

```python
similar
```

```python
sorted(list(enumerate(similar[0])),reverse=True,key = lambda x:x[1])
```

```python
def recommend(movie):

  index = movies[movies['title']==movie].index[0]

  distance = similar[index]

  movies_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x:x[1])[1:6]
  for i in movies_list:

    print(movies.iloc[i[0]].title)
```

```python
recommend("Trial By Media")
```
